## Data Preprocessing and Adding Feature

```python
# Add TotalAcceptedCamp Column
df["TotalAcceptedCmp"] = df['AcceptedCmp1'] + df['AcceptedCmp2'] + df['AcceptedCmp3'] + df['AcceptedCmp4'] + df['AcceptedCmp5']

# Add Monetary Column
df["Monetary"] = df['MntWines'] + df['MntFruits'] + df['MntMeatProducts'] + df['MntFishProducts'] + df['MntSweetProducts'] + df['MntGoldProds']

# Add Frequency Column
df["Frequency"] = df['NumDealsPurchases'] + df['NumWebPurchases'] + df['NumCatalogPurchases'] + df['NumStorePurchases'] 
```

```python
df[df['Frequency']==0]
```

```python
# Drop Customer with 0 Shopping Frequency

df.drop(df[df['Frequency'] == 0].index,inplace=True)
```

```python
# Drop unused columns 'Z_CostContact' and 'Z_Revenue'

df = df.drop(columns=['Z_CostContact', 'Z_Revenue'], axis=1)
```

The Enrollment_Date feature was added with a datetime type to facilitate date operations, enhance trend analysis, and improve data accuracy, as the original Dt_Customer column was of object data type.

```python
df['Enrollment_Date'] = pd.to_datetime(df['Dt_Customer'], format='%d-%m-%Y')

df[['ID', 'Dt_Customer', 'Enrollment_Date']].head()
```

```python
min_date = df['Enrollment_Date'].min()
max_date = df['Enrollment_Date'].max()

print("Minimum Enrollment Date:", min_date)
print("Maximum Enrollment Date:", max_date)
```

The year 2016 was used as the current year because after converting Dt_Customer, the most recent customer enrollment was identified as June 29, 2014. Given that the products spend data covers the last two years, 2016 accurately reflects the analysis period.

```python
# Add Age_Range Column

current_year = 2016
df['Age'] = current_year - df['Year_Birth']

bins = [0, 18, 25, 35, 45, 55, 65, 100]
labels = ['0-17', '18-24', '25-34', '35-44', '45-54', '55-64', '65+']
df['Age_Range'] = pd.cut(df['Age'], bins=bins, labels=labels)

df[['ID', 'Year_Birth', 'Age', 'Age_Range']].head()
```

```python
# Add Income_Category Column

income_category = {
    "Low Income": df['Income'].quantile(0.25),           
    "Lower Middle Income": df['Income'].quantile(0.50),  
    "Upper Middle Income": df['Income'].quantile(0.75),   
}

def categorize_income(income):
    if income <= income_category["Low Income"]:
        return "Low Income"
    elif income <= income_category["Lower Middle Income"]:
        return "Lower Middle Income"
    elif income <= income_category["Upper Middle Income"]:
        return "Upper Middle Income"
    else:
        return "High Income"

df['Income_Category'] = df['Income'].apply(categorize_income)

df[['ID', 'Income', 'Income_Category']].head()
```

```python
# Add Education_Group

education_mapping = {
    'Basic': 'Diploma',
    'Graduation': 'Bachelor',
    'Master': 'Master',
    '2n Cycle': 'Master',
    'PhD': 'Doctoral'
}

df['Education_Group'] = df['Education'].replace(education_mapping)

df[['ID', 'Education', 'Education_Group']].head()
```

```python
# Add Preferred_Channels

df.loc[(df['NumWebPurchases'] > df['NumCatalogPurchases']) & (df['NumWebPurchases'] > df['NumStorePurchases']), 'Preferred_Channels'] = 'Website'
df.loc[(df['NumStorePurchases'] > df['NumCatalogPurchases']) & (df['NumStorePurchases'] > df['NumWebPurchases']), 'Preferred_Channels'] = 'Offline Store'
df.loc[(df['NumCatalogPurchases'] > df['NumStorePurchases']) & (df['NumCatalogPurchases'] > df['NumWebPurchases']), 'Preferred_Channels'] = 'Catalog'
df.loc[((df['NumWebPurchases'] == df['NumCatalogPurchases']) & (df['NumWebPurchases'] > df['NumStorePurchases'])) |
       ((df['NumWebPurchases'] == df['NumStorePurchases']) & (df['NumWebPurchases'] > df['NumCatalogPurchases'])) |
       ((df['NumCatalogPurchases'] == df['NumStorePurchases']) & (df['NumCatalogPurchases'] > df['NumWebPurchases'])) |
       ((df['NumWebPurchases'] == df['NumCatalogPurchases']) & (df['NumWebPurchases'] == df['NumStorePurchases'])), 'Preferred_Channels'] = 'No Preference'

df[['ID', 'Preferred_Channels']].head()
```

```python
# Add Marital_Status_Group Column

marital_status_mapping = {
    'Single': 'Single',
    'Together': 'Partnered',
    'Married': 'Married',
    'Divorced': 'Divorced',
    'Widow': 'Divorced',
    'Alone': 'Single',
    'Absurd': 'Single',
    'YOLO': 'Single'
}

df['Marital_Status_Group'] = df['Marital_Status'].replace(marital_status_mapping)

df[['ID', 'Marital_Status', 'Marital_Status_Group']].head()
```

```python
# Add Family_Size Column

def calculate_family_size(row):
    if row['Marital_Status_Group'] in ['Single', 'Divorced']:
        base_size = 1
    elif row['Marital_Status_Group'] in ['Partnered', 'Married']:
        base_size = 2
    else:
        base_size = 0

    return base_size + row['Kidhome'] + row['Teenhome']

df['Family_Size'] = df.apply(calculate_family_size, axis=1)

df[['ID', 'Marital_Status_Group', 'Kidhome', 'Teenhome', 'Family_Size']].head()
```

## Problem 2: How does customer demographics affect spending patterns across different product categories in supermarkets?

The analysis seeks to uncover how customer demographics affect spending patterns across various product categories in supermarkets, focusing on wines, fruits, meat products, fish products, sweet products, and gold products. By examining relationships between demographic factorsâ€”such as age, income, education level, marital status, and family sizeâ€”and spending behaviors, the study aims to uncover trends that can inform supermarket strategies.

```python
product_categories = ['MntWines', 'MntFruits', 'MntMeatProducts', 
                      'MntFishProducts', 'MntSweetProducts', 'MntGoldProds']

avg_spending_by_age = df.groupby('Age_Range', observed=False)[product_categories].mean().reset_index()
avg_spending_by_education = df.groupby('Education_Group', observed=False)[product_categories].mean().reset_index()
avg_spending_by_income = df.groupby('Income_Category', observed=False)[product_categories].mean().reset_index()
avg_spending_by_marital_status = df.groupby('Marital_Status_Group', observed=False)[product_categories].mean().reset_index()
avg_spending_by_family_size = df.groupby('Family_Size', observed=False)[product_categories].mean().reset_index()


fig, axes = plt.subplots(5, 1, figsize=(15, 30))

# Plot for Age Range
avg_spending_by_age.plot(kind='bar', x='Age_Range', ax=axes[0])
axes[0].set_title('Average Spending by Age Group')
axes[0].set_xlabel('Age Group')
axes[0].set_ylabel('Average Spending')
axes[0].legend(title='Product Categories')
axes[0].set_xticklabels(avg_spending_by_age['Age_Range'], rotation=0)
for p in axes[0].patches:
    height = round(p.get_height())
    if height != 0:
        axes[0].annotate(f'{height}', (p.get_x() + p.get_width() / 2, p.get_height() + 1),
                         ha='center', va='bottom', fontsize=8, color='black', fontweight='bold')
        

# Plot for Education Level
avg_spending_by_education.plot(kind='bar', x='Education_Group', ax=axes[1])
axes[1].set_title('Average Spending by Education Level')
axes[1].set_xlabel('Education Level')
axes[1].set_ylabel('Average Spending')
axes[1].legend(title='Product Categories')
axes[1].set_xticklabels(avg_spending_by_education['Education_Group'], rotation=0)
for p in axes[1].patches:
    height = round(p.get_height())
    if height != 0:
        axes[1].annotate(f'{height}', (p.get_x() + p.get_width() / 2, p.get_height() + 1),
                         ha='center', va='bottom', fontsize=8, color='black', fontweight='bold')

# Plot for Income Category
avg_spending_by_income.plot(kind='bar', x='Income_Category', ax=axes[2])
axes[2].set_title('Average Spending by Income Category')
axes[2].set_xlabel('Income Category')
axes[2].set_ylabel('Average Spending')
axes[2].legend(title='Product Categories')
axes[2].set_xticklabels(avg_spending_by_income['Income_Category'], rotation=0)
for p in axes[2].patches:
    height = round(p.get_height())
    if height != 0:
        axes[2].annotate(f'{height}', (p.get_x() + p.get_width() / 2, p.get_height() + 1),
                         ha='center', va='bottom', fontsize=8, color='black', fontweight='bold')

# Plot for Marital Status
avg_spending_by_marital_status.plot(kind='bar', x='Marital_Status_Group', ax=axes[3])
axes[3].set_title('Average Spending by Marital Status')
axes[3].set_xlabel('Marital Status')
axes[3].set_ylabel('Average Spending')
axes[3].legend(title='Product Categories')
axes[3].set_xticklabels(avg_spending_by_marital_status['Marital_Status_Group'], rotation=0)
for p in axes[3].patches:
    height = round(p.get_height())
    if height != 0:
        axes[3].annotate(f'{height}', (p.get_x() + p.get_width() / 2, p.get_height() + 1),
                         ha='center', va='bottom', fontsize=8, color='black', fontweight='bold')

# Plot for Family Size
avg_spending_by_family_size.plot(kind='bar', x='Family_Size', ax=axes[4])
axes[4].set_title('Average Spending by Family Size')
axes[4].set_xlabel('Family Size')
axes[4].set_ylabel('Average Spending')
axes[4].legend(title='Product Categories')
axes[4].set_xticklabels(avg_spending_by_family_size['Family_Size'], rotation=0)
for p in axes[4].patches:
    height = round(p.get_height())
    if height != 0:
        axes[4].annotate(f'{height}', (p.get_x() + p.get_width() / 2, p.get_height() + 1),
                         ha='center', va='bottom', fontsize=8, color='black', fontweight='bold')

plt.tight_layout()
plt.show()
```

### Insight Point of Spending Pattern from Product Categories


**1. Product Categories by Age Group**

Observations:
The data shows that middle-aged shoppers (35-44) spend the most across all product categories. ## Problem 3: How effective are the marketing campaigns in reaching and engaging customers?

This analysis is conducted to measure the effectiveness of Campaigns 1 through 5 and to understand how customers income levels impact their RFM (Recency, Frequency, and Monetary) metrics. The goal is to evaluate how well each campaign engages customers and whether income influences how frequently and recently customers shop, as well as how much they spend. By examining both campaign performance and the effect of income on shopping behaviors, we can gain insights into which customer segments are most responsive to marketing efforts and identify areas for improvement.

```python
def campaign_acceptance_rates(df):
    campaigns = ['AcceptedCmp1', 'AcceptedCmp2', 'AcceptedCmp3', 'AcceptedCmp4', 'AcceptedCmp5']
    
    for campaign in campaigns:
        if campaign not in df.columns:
            print(f"Warning: {campaign} is not in the DataFrame.")
    
    acceptance_rates = df[campaigns].mean() * 100
    
    never_accepted = (df[campaigns].sum(axis=1) == 0).mean() * 100
    acceptance_rates["Never Accepted"] = never_accepted
    
    return acceptance_rates


acceptance_rates = campaign_acceptance_rates(df)


plt.figure(figsize=(6, 6))
plt.pie(
    acceptance_rates, 
    labels=acceptance_rates.index, 
    autopct='%1.1f%%', 
    startangle=140, 
    colors=sns.color_palette('viridis', len(acceptance_rates))
)
plt.title('Marketing Campaign Acceptance Rates')
plt.axis('equal') 
plt.tight_layout()
plt.show()
```

### Recency, Frequency, Monetary (RFM) Analysis

RFM analysis segments customers based on three factors: Recency (how recently a customer made a purchase), Frequency (how often they make purchases), and Monetary Value (how much they spend). By scoring customers on each metric, RFM helps businesses identify and prioritize high-value customers, allowing for targeted marketing strategies. For instance, customers who score high on all three metrics are often the most loyal and profitable, making them ideal targets for special offers or rewards. RFM analysis helps businesses boost customer retention, improve loyalty, and maximize revenue by focusing on the customers who are most likely to engage and spend.

```python
from sklearn.linear_model import LinearRegression

hue_vars = ['AcceptedCmp1', 'AcceptedCmp2', 'AcceptedCmp3', 'AcceptedCmp4', 'AcceptedCmp5']
colors = ['red', 'blue', 'green', 'purple', 'orange']


# SCATTERPLOT 1: Income vs Recency with Linear Regression Line
reg2 = LinearRegression().fit(df[['Recency']], df['Income'])
df['bestfit1'] = reg2.predict(df[['Recency']])

plt.figure(figsize=(8, 5))
for hue, color in zip(hue_vars, colors):
    filtered_df = df[df[hue] == 1]
    plt.scatter(filtered_df['Recency'], filtered_df['Income'], label=f'Accepted {hue}', color=color, alpha=0.6)
plt.plot(df['Recency'], df['bestfit1'], color='black', label='Line of Best Fit')
plt.xlabel('Recency')
plt.ylabel('Income')
plt.title('Income to Recency by Campaign Acceptance')
plt.legend()
plt.show()

# SCATTERPLOT 2: Income vs Frequency with Linear Regression Line
reg3 = LinearRegression().fit(df[['Frequency']], df['Income'])
df['bestfit2'] = reg3.predict(df[['Frequency']])

plt.figure(figsize=(8, 5))
for hue, color in zip(hue_vars, colors):
    filtered_df = df[df[hue] == 1]
    plt.scatter(filtered_df['Frequency'], filtered_df['Income'], label=f'Accepted {hue}', color=color, alpha=0.6)
plt.plot(df['Frequency'], df['bestfit2'], color='black', label='Line of Best Fit')
plt.xlabel('Frequency')
plt.ylabel('Income')
plt.title('Income to Frequency by Campaign Acceptance')
plt.legend()
plt.show()

# SCATTERPLOT 3: Income vs Monetary with Linear Regression Line
reg1 = LinearRegression().fit(df[['Monetary']], df['Income'])
df['bestfit3'] = reg1.predict(df[['Monetary']])

plt.figure(figsize=(8, 5))
for hue, color in zip(hue_vars, colors):
    filtered_df = df[df[hue] == 1]
    plt.scatter(filtered_df['Monetary'], filtered_df['Income'], label=f'Accepted {hue}', color=color, alpha=0.6)
plt.plot(df['Monetary'], df['bestfit3'], color='black', label='Line of Best Fit')
plt.xlabel('Monetary')
plt.ylabel('Income')
plt.title('Income to Monetary by Campaign Acceptance')
plt.legend()
plt.show()
```

## RFM Analysis Insight Point

**1. Income to Recency by Campaign Acceptance:**
The scatterplot shows no association correlation between income and recent purchase activity, as data points appear scattered. Customers across income levels exhibit similar recent purchase behaviors, indicating that income alone doesn't predict recency. These insights imply that, while income does not directly correlate with recency, certain campaigns resonate more effectively with particular customer profiles.

The scatterplot shows the relationship between income and purchase frequency reveals a moderate positive linear correlation between the two variables. This means that, as income increases, the frequency of purchases tends to rise as well. The data points appear to follow an upward trend, suggesting that customers with higher income levels are likely to make more frequent purchases. This observation indicates that income may play a significant role in influencing consumer buying behavior, where those with greater financial resources engage in shopping more often.

**3. Income to Monetary by Campaign Acceptance:**
The scatterplot show the relationship between income and monetary value reveals a strong positive linear correlation between the two variables, indicating that as income increases, monetary value tends to rise as well. The points generally cluster around a straight line, reflecting a clear upward trend; however, the relationship is moderately strong, as evidenced by considerable scatter around the line of best fit, which suggests that other factors may also influence monetary value.
